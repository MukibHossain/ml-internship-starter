# 2. Artificial Intelligence (AI)

## Definition

Artificial Intelligence (AI) is the broad field of computer science that focuses on building systems capable of performing tasks that normally require human intelligence, such as reasoning, learning, problem-solving, decision-making, and language understanding.

## Applications

- Self-driving cars
- Virtual assistants
- Medical diagnosis
- Robotics
- Language translation

## Example

A chatbot that understands user questions and responds intelligently.

# 3. Machine Learning (ML)

## Definition

Machine Learning (ML) is a subset of Artificial Intelligence that enables computers to learn patterns from data and make predictions or decisions without being explicitly programmed.

## Types

- Supervised Learning
- Unsupervised Learning
- Reinforcement Learning

## Applications

- Disease prediction
- Fraud detection
- Spam email filtering
- Image recognition
- Recommendation systems

## Example

Predicting whether a patient has diabetes based on medical information.

# 4. Analytics

## Definition

Analytics is the process of collecting, analyzing, and interpreting data to support decision-making.

## Types of Analytics

### Descriptive Analytics
Answers: **What happened?**

Example:
Monthly sales report.

### Diagnostic Analytics
Answers: **Why did it happen?**

Example:
Analyzing reasons for decreased sales.

### Predictive Analytics
Answers: **What is likely to happen?**

Example:
Predicting customer churn.

### Prescriptive Analytics
Answers: **What should we do?**

Example:
Recommending the best marketing strategy.

# 5. Rule-Based Systems

## Definition

Rule-based systems solve problems using predefined IF–THEN rules rather than learning from data.

## Advantages

- Easy to understand
- Fast execution
- No training data required
- Predictable behavior

## Limitations

- Cannot learn new patterns
- Difficult to maintain with many rules
- Poor performance on complex problems

## Example

IF Age ≥ 18

THEN Eligible to Vote

# 6. AI vs ML vs Analytics vs Rule-Based Systems

- **Artificial Intelligence (AI):** A broad field focused on creating systems that can perform tasks requiring human intelligence.
- **Machine Learning (ML):** A subset of AI that learns patterns from data to make predictions or decisions.
- **Analytics:** The process of examining data to gain insights and support decision-making.
- **Rule-Based Systems:** Systems that follow predefined IF–THEN rules without learning from data.

# 7. When Should We Use Machine Learning?

Machine Learning should be used when:

- Large amounts of data are available.
- Relationships between variables are complex.
- The problem requires prediction.
- Rules cannot capture all possible patterns.

## Examples

- Disease prediction
- Face recognition
- Customer recommendation
- Credit risk prediction
- Traffic forecasting
