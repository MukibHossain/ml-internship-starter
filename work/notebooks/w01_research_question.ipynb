## 2. The question: decision, action, cost of a wrong call

### Research Question
Which search and content signals are most strongly associated with better content performance and should be prioritized for review?

### Decision
The decision is which content pages should be reviewed or improved first based on their performance signals.

### Unit of Analysis
One content page (one row represents one page in the dataset).

### Output
A ranked list of the most important signals affecting content performance and recommendations based on those signals.

### Action
A content or SEO team can use these findings to prioritize which pages to update, optimize, or monitor.

### Cost of a Wrong Recommendation
If the analysis incorrectly identifies important signals, the team may spend time improving the wrong pages while missing pages that actually need attention. This could reduce the effectiveness of optimization efforts and waste resources.

### Why Machine Learning or Data Analysis Helps
The dataset contains many pages and multiple performance metrics. Data analysis and machine learning can identify patterns across all pages that would be difficult to detect manually.
